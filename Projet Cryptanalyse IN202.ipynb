```python
def frequence_lettres(message_chiffre):
    """ donne la fréquence des lettres dans un texte"""
    resultat = []
    taille = len(message_chiffre) - message_chiffre.count(' ')
    for lettre in message_chiffre:
        if 97 <= ord(lettre) <= 122:
            exist = True
            for i in range(len(resultat)):
                if resultat[i][0] == lettre:
                    exist = False 
            if exist:
                resultat.append([lettre, round(message_chiffre.count(lettre)/taille*100,2)])
    return sorted(resultat, key=lambda pair: pair[1], reverse=True)
```

```python
def rang(lettre):
    """ permet de numéroter les lettres de l'alphabet de 0 à 25 """
    return ord(lettre) - 97 

def decalage(lettre_message, lettre_cle):
    """" décale la lettre_message du nombre correspondant à lettre_clé"""
    if 97 <= ord(lettre_message) <= 122:
        return chr((rang(lettre_message) + rang(lettre_cle))%26 + 97)
    else:
        return lettre_message
```

```python
def cle_inverse(cle):
    """ renvoie la l'inverse de la clé permettant de decoder un texte"""
    res = ""
    for elem in cle: 
        if elem == 'a':
            res += 'a'
        else :
            res += chr(26-rang(elem)+97)
    return res
```

```python
def cle_hyp(depart, arrivee):
    """ renvoie la clé permettant de passer du mot de départ à celui d'arrivé"""
    res = ''
    for i in range(len(depart)):
        d = rang(depart[i])
        a = rang(arrivee[i])
        if d <= a :
            res += chr(a-d+97)
        else :
            res += chr(26-(d-a)+97)
    return res
```

```python
def dec_texte(texte,cle):
    """ renvoie le texte modifié à l'aide de la clé """
    taille_cle = len(cle)
    res = ""
    for i in range(len(texte)):
        res += decalage(texte[i],cle[i%taille_cle])
    return res

def chiffre():
    resultat.delete(0,tk.END)
    text=entree_texte.get()
    cle=entree_cle.get()
    if((len(text)>0) and (len(cle)>0)):
        res=dec_texte(text,cle)
        resultat.insert(0,res)
    else:
        resultat.insert(0,"Il manque quelque chose")
```

```python
def comparaison(texte):
    """ renvoie une liste donnant la correspondance entre les lettres du texte codé et celles 'normales' selon la fréquence"""
    res = []
    frequence_texte = frequence_lettres(texte)
    for i in range(len(frequence)):
        if i >= len(frequence_texte):
            res.append(['-', frequence[i]])
        else:
            res.append([frequence_texte[i][0], frequence[i]])
    return res
```

```python
def traduction(texte):
    """ renvoie la chaine de caractère avec les lettres échangées si cc est mofifié"""
    res = ""
    for lettre in texte:
        res += lettre
        for elem in cc:
            if lettre == elem[0]:
                res = res[:-1]
                res += elem[1]
                break
    return res
```

```python
def changer_comparaison_lettre(entree1, entree2):
    """ echange la place de deux lettres dans cc """
    for j in range(len(cc)):
        if cc[j][1] == entree1:
            variable = cc[j][0]
            cc.remove(cc[j])
            cc.insert(j, [variable, entree2])
        elif cc[j][1] == entree2:
            variable = cc[j][0]
            cc.remove(cc[j])
            cc.insert(j, [variable, entree1])
```

```python
def changer_comparaison_mots(mot1, mot2):
    """ échange la place de plusieurs lettres dans cc """
    if len(mot1) == 0:
        changer_comparaison_lettre(mot1, mot2)
    else : 
        trad = str()
        for i in range(len(mot1)):
            for j in range(len(cc)):
                if mot1[i] == cc[j][1]:
                    trad += cc[j][0]
        for i in range(len(mot1)):
            mot1 = traduction(trad)
            changer_comparaison_lettre(mot1[i], mot2[i])
    return traduction(texte_2)
```

```python
def mots_n_lettres(n, texte):
    """ renvoie une liste contenant tous les mots à n lettres d'un texte et leur position"""
    res = []
    index = 0
    while index < len(texte)-n:
        sous_res = ''
        for i in range(n):
            if not 0 <= rang(texte[index+i]) <= 26:
                break
            else : 
                sous_res += texte[index+i]
                if len(sous_res) == n:
                    if index == 0 and texte[index+n] == ' ':
                        res.append([sous_res, index])
                    elif texte[index-1] == ' ' and texte[index+n] == ' ':
                        res.append([sous_res, index])
        index += 1
    return res 
```

```python
def condense(texte):
    """ renvoie le texte composé seulement de lettres"""
    res = ""
    for elem in texte:
        if 0 <= rang(elem) <= 26:
            res += elem
    return res
```

```python
def hypotheses_cles(cle_taille, text):
    """ à l'aide de nos hypothèses, on essaye de trouver les clés possibles à clé_taille lettre """

    aide_a = mots_n_lettres(1, text) 
    aide_l = ['k', 83]
    aide_e = ['g', 84]
    txt_lettres = condense(text)
    cle = ''

    for i in range(cle_taille):
        repet = 0
        cle_lettre = 0

        while repet*cle_taille + i < len(txt_lettres):

            for lettre in aide_a:
                if i+repet*cle_taille == lettre[1] and cle_lettre == 0:
                    cle_lettre = cle_hyp('a',lettre[0])

            if i+repet*cle_taille == aide_e[1] and cle_lettre == 0:
                    cle_lettre = cle_hyp('e',aide_e[0])

            if i+repet*cle_taille == aide_l[1] and cle_lettre == 0:
                    cle_lettre = cle_hyp('l',aide_l[0])

            repet += 1

        if cle_lettre == 0:
            cle += '?'
        else : 
            cle += cle_lettre

    return cle 
```

```python
def alphabet():
    res = []
    for i in range(26):
        res.append(chr(i+97))
    return res
```

```python
def choix_cle(hyp_cle, texte):
    """ trouve la lettre manquante de cle (s'il n'y en a qu'une) """
    res = []
    let = alphabet()
    for lettre in let:
        nv_cle = ''
        for elem in hyp_cle:
            if elem == '?':
                nv_cle += lettre
            else:
                nv_cle += elem
        if frequence_lettres(dec_texte(texte,cle_inverse(nv_cle)))[0][1] > 17.5:
            res.append([nv_cle, cle_inverse(nv_cle)])
    return res
```

```python
def taille_cle(n, texte):
    """ renvoie n chaine de caractère pour une clé de n lettres"""
    res = []
    for i in range(n):
        sous_res = ""
        count = 0
        while count*n + i < len(texte):
            sous_res += texte[i + count*n]
            count += 1
        res.append(sous_res)
    return res 
```

```python
def cle_poss(n, texte):
    """ renvoie les valeurs possibles de la première à la n-ième lettre de la clé"""
    res = ''
    res_freq = []
    for elem in taille_cle(n, texte):
        frequence = frequence_lettres(elem)
        sous_res = []
        for i in range(len(frequence)):
            if frequence[i][1] == frequence[0][1]:
                sous_res.append(cle_hyp('e', frequence[i][0]))
        res_freq.append(sous_res)
    return res_freq
```

En suivant les indices du texte 3, on suppose qu'il faut remettre chaque phrase à l'endroit. Je vais différencier deux cas : 1. chaque phrase est inversée et on ne garde pas le formatage  2. chaque phrase est inversée et on garde le formatage 

```python
def inverser_sens(nom_fichier):
    """ enlève le formatage et renvoie les phrases lues de droite à gauche"""
    fichier = open(nom_fichier, 'r')
    res1 = ""
    with fichier as f:
        for line in f.readlines():
            for i in range(1, len(line)):
                res1 += line[len(line)-1-i]
            res1 += ' '
    fichier.close()
    return res1
```

```python
def inverser_sens_ligne(nom_fichier):
    """ garde le formatage et renvoie les phrases de droites à gauches"""
    fichier = open(nom_fichier, 'r')
    lignes = []
    res = []
    with fichier as f:
        for line in f.readlines():
            lignes.append(line)
    for line in lignes:
        res2 = ""
        for i in range(1, len(line)):
            res2 += line[len(line)-1-i]
        res.append(res2)
    res.remove('')

    fichier.close()
    return res
```

```python
def groupe_n_mots(n, texte):
    """renvoie la position et toutes les répétitions de n lettres dans une chaine de caractères"""
    res = []
    res2 = []
    res3 = []
    res4 = []
    for elem in texte:
        for j in range(len(elem)):
            compare = elem[j:j+n]
            res.append(compare)
            res3.append([j, compare])
    for i in range(len(res)):
        if res.count(res[i]) > 1 and len(res[i]) == n: 
            res4.append(res3[i])
            res2.append(res[i])
    return res4
```

```python
def groupe_n_mots_2(n, texte):
    """ même chose mais pour toutes les chaînes de caractères d'une liste"""
    res = []
    res1 = []
    res2 = []
    for j in range(len(texte)):
        compare = texte[j:j+n]
        res.append([j, compare])
        res1.append(compare)
    for i in range(len(res1)):
        if res1.count(res1[i]) > 1:
            res2.append(res[i])
    return res2
```
